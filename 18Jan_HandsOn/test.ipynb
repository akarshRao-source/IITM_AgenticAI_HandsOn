```python
import csv

def calculate_emi(principal, annual_rate, tenure_months):
    monthly_rate = annual_rate / (12 * 100)

    if monthly_rate == 0:
        return principal / tenure_months

    emi = principal * monthly_rate * (1 + monthly_rate) ** tenure_months / \
          ((1 + monthly_rate) ** tenure_months - 1)
    return emi


def generate_amortization_table(principal, annual_rate, tenure_months, emi):
    monthly_rate = annual_rate / (12 * 100)
    balance = principal

    table = []
    total_interest = 0

    for month in range(1, tenure_months + 1):
        interest = balance * monthly_rate
        principal_paid = emi - interest
        balance -= principal_paid

        total_interest += interest

        table.append([
            month,
            round(emi, 2),
            round(principal_paid, 2),
            round(interest, 2),
            round(balance if balance > 0 else 0, 2)
        ])

    return table, total_interest


def export_to_csv(table, filename="amortization_schedule.csv"):
    headers = [
        "Month",
        "EMI",
        "Principal Paid",
        "Interest Paid",
        "Remaining Balance"
    ]

    with open(filename, mode="w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(headers)
        writer.writerows(table)

    print(f"\nCSV file exported successfully: {filename}")


def display_table(table, rows=12):
    print("\nMonth | EMI | Principal | Interest | Balance")
    print("-" * 55)

    for row in table[:rows]:
        print(
            f"{row[0]:5} | {row[1]:8,.2f} | {row[2]:9,.2f} | "
            f"{row[3]:8,.2f} | {row[4]:10,.2f}"
        )

    if len(table) > rows:
        print("... (remaining months omitted)")


def tenure_comparison(principal, annual_rate, tenures_years):
    print("\n====== Tenure Comparison ======")
    print("Years | EMI | Total Payment | Total Interest")
    print("-" * 60)

    for years in tenures_years:
        months = years * 12
        emi = calculate_emi(principal, annual_rate, months)
        total_payment = emi * months
        total_interest = total_payment - principal

        print(
            f"{years:5} | {emi:8,.2f} | {total_payment:13,.2f} | {total_interest:13,.2f}"
        )


def main():
    print("====== Loan Calculator (Amortization + Comparison) ======")

    principal = 100000 # float(input("Enter loan amount (Principal): "))
    annual_rate = 12 #float(input("Enter annual interest rate (%): "))

    tenure_type = "Y" # input("Is tenure in Years or Months? (Y/M): ").strip().upper()
    tenure_value = 12 #int(input("Enter loan tenure: "))

    if tenure_type == "Y":
        tenure_months = tenure_value * 12
    elif tenure_type == "M":
        tenure_months = tenure_value
    else:
        print("Invalid tenure type.")
        return

    emi = calculate_emi(principal, annual_rate, tenure_months)

    table, total_interest = generate_amortization_table(
        principal, annual_rate, tenure_months, emi
    )

    print("\n====== Loan Summary ======")
    print(f"Monthly EMI         : ₹{emi:,.2f}")
    print(f"Total Payment       : ₹{emi * tenure_months:,.2f}")
    print(f"Total Interest Paid : ₹{total_interest:,.2f}")

    # Display first few rows
    display_table(table)

    # Export CSV
    export_to_csv(table)

    # Tenure comparison
    tenure_comparison(
        principal,
        annual_rate,
        tenures_years=[10, 15, 20, 25]
    )


if __name__ == "__main__":
    main()
```

```python
import gradio as gr

def format_tasks(tasks):
    if not tasks:
        return "📭 No tasks yet."

    return "\n".join(
        f"{i+1}. {task[0]} — {'✅ Completed' if task[1] else '⏳ Pending'}"
        for i, task in enumerate(tasks)
    )


def get_task_names(tasks):
    return [task[0] for task in tasks]


def add_task(task_name, tasks):
    if not task_name.strip():
        return tasks, "❌ Task cannot be empty.", format_tasks(tasks), gr.update(choices=[]), ""

    tasks.append([task_name, False])
    return tasks, "✅ Task added successfully.", format_tasks(tasks), gr.update(choices=get_task_names(tasks)), ""


def mark_completed(selected_task, tasks):
    if not selected_task:
        return tasks, "❌ Please select a task.", format_tasks(tasks), gr.update(choices=get_task_names(tasks))

    for task in tasks:
        if task[0] == selected_task:
            task[1] = True
            return tasks, f"✅ '{selected_task}' marked as completed.", format_tasks(tasks), gr.update(choices=get_task_names(tasks))


def delete_task(selected_task, tasks):
    if not selected_task:
        return tasks, "❌ Please select a task.", format_tasks(tasks), gr.update(choices=get_task_names(tasks))

    tasks[:] = [task for task in tasks if task[0] != selected_task]
    return tasks, f"🗑️ '{selected_task}' deleted.", format_tasks(tasks), gr.update(choices=get_task_names(tasks))


def move_task_up(selected_task, tasks):
    index = next((i for i, t in enumerate(tasks) if t[0] == selected_task), None)
    if index is None or index == 0:
        return tasks, "❌ Cannot move up.", format_tasks(tasks), gr.update(choices=get_task_names(tasks))

    tasks[index - 1], tasks[index] = tasks[index], tasks[index - 1]
    return tasks, "⬆ Task moved up.", format_tasks(tasks), gr.update(choices=get_task_names(tasks))


def move_task_down(selected_task, tasks):
    index = next((i for i, t in enumerate(tasks) if t[0] == selected_task), None)
    if index is None or index == len(tasks) - 1:
        return tasks, "❌ Cannot move down.", format_tasks(tasks), gr.update(choices=get_task_names(tasks))

    tasks[index + 1], tasks[index] = tasks[index], tasks[index + 1]
    return tasks, "⬇ Task moved down.", format_tasks(tasks), gr.update(choices=get_task_names(tasks))


with gr.Blocks(title="To-Do List App") as app:
    gr.Markdown("## 📝 To-Do List App")

    tasks_state = gr.State([])

    # --- Add Task ---
    gr.Markdown("### ➕ Add Task")
    task_input = gr.Textbox(label="Task Description")
    add_btn = gr.Button("Add Task")

    # --- Mark Completed ---
    gr.Markdown("### ✅ Mark Completed")
    task_dropdown = gr.Dropdown(label="Select Task", choices=[])
    complete_btn = gr.Button("Mark as Completed")

    # --- Delete Task ---
    gr.Markdown("### 🗑️ Delete Task")
    delete_btn = gr.Button("Delete Task")

    # --- Reorder Tasks ---
    gr.Markdown("### 🔀 Reorder Priority")
    up_btn = gr.Button("⬆ Move Up")
    down_btn = gr.Button("⬇ Move Down")

    message = gr.Markdown()
    task_display = gr.Textbox(label="Your Tasks", lines=10, interactive=False)

    add_btn.click(
        add_task,
        [task_input, tasks_state],
        [tasks_state, message, task_display, task_dropdown, task_input]
    )

    complete_btn.click(
        mark_completed,
        [task_dropdown, tasks_state],
        [tasks_state, message, task_display, task_dropdown]
    )

    delete_btn.click(
        delete_task,
        [task_dropdown, tasks_state],
        [tasks_state, message, task_display, task_dropdown]
    )

    up_btn.click(
        move_task_up,
        [task_dropdown, tasks_state],
        [tasks_state, message, task_display, task_dropdown]
    )

    down_btn.click(
        move_task_down,
        [task_dropdown, tasks_state],
        [tasks_state, message, task_display, task_dropdown]
    )

app.launch()
```

```python
def split_expense(total, names, weights=None, round_to=2):
    """
    Splits a total expense among people, supporting uneven shares via weights,
    with strong type validation.
    """

    # --- Type checks ---
    if not isinstance(total, (int, float)):
        raise TypeError("Total must be a number (int or float).")

    if not isinstance(names, list):
        raise TypeError("Names must be provided as a list.")

    if not all(isinstance(name, str) for name in names):
        raise TypeError("Each name must be a string.")

    if weights is not None and not isinstance(weights, (list, dict)):
        raise TypeError("Weights must be a list or a dictionary.")

    if not isinstance(round_to, int):
        raise TypeError("round_to must be an integer.")

    # --- Value validation ---
    if total < 0:
        raise ValueError("Total amount cannot be negative.")

    if not names:
        raise ValueError("Names list cannot be empty.")

    if round_to < 0:
        raise ValueError("round_to must be zero or positive.")

    # --- Default equal weights ---
    if weights is None:
        weights = {name: 1 for name in names}

    # --- Convert weights list to dict ---
    if isinstance(weights, list):
        if len(weights) != len(names):
            raise ValueError("Weights list must match names list length.")
        if not all(isinstance(w, (int, float)) for w in weights):
            raise TypeError("All weights must be numbers.")
        weights = dict(zip(names, weights))

    # --- Validate weights dict ---
    if isinstance(weights, dict):
        for name, weight in weights.items():
            if not isinstance(weight, (int, float)):
                raise TypeError(f"Weight for {name} must be a number.")
            if weight <= 0:
                raise ValueError("Weights must be positive numbers.")

    # --- Core calculation ---
    total_weight = sum(weights.values())

    split = {}
    for name in names:
        share = (weights[name] / total_weight) * total
        split[name] = round(share, round_to)

    return split


########################

total = 500
names = ["Alice", "Bob", "Charlie"]
weights = {"Alice": 1, "Bob": 2, "Charlie": 1}

result = split_expense(total, names, weights=weights, round_to=2)
print(result)

########################
total = 1000
names = ["Rahul", "Anita", "Vikram"]
weights = {"Rahul": 2, "Anita": 1, "Vikram": 1}
result = split_expense(total, names, weights=weights, round_to=2)
print(result)
########################
total = 750
names = ["Solo"]
result = split_expense(total, names)
print(result)


#########################
# ✅ Error Handling Demo - Invalid Types
print("\n--- Error Handling Examples ---\n")

# Test Case 1: Invalid input types (string instead of number and list)
try:
    print("Test 1: Invalid types (total='rahul', names='Anita')")
    total = "rahul" 
    names = "Anita"
    result = split_expense(total, names)
    print(result)
except TypeError as e:
    print(f"✅ TypeError caught: {e}\n")
except ValueError as e:
    print(f"✅ ValueError caught: {e}\n")

# Test Case 2: Negative total
try:
    print("Test 2: Negative total amount")
    total = -500
    names = ["Alice", "Bob"]
    result = split_expense(total, names)
    print(result)
except ValueError as e:
    print(f"✅ ValueError caught: {e}\n")

# Test Case 3: Empty names list
try:
    print("Test 3: Empty names list")
    total = 1000
    names = []
    result = split_expense(total, names)
    print(result)
except ValueError as e:
    print(f"✅ ValueError caught: {e}\n")

# Test Case 4: Invalid weights (negative value)
try:
    print("Test 4: Negative weight values")
    total = 1000
    names = ["Alice", "Bob"]
    weights = {"Alice": -5, "Bob": 2}
    result = split_expense(total, names, weights=weights)
    print(result)
except ValueError as e:
    print(f"✅ ValueError caught: {e}\n")

# Test Case 5: Mixed valid and invalid scenario
try:
    print("Test 5: Type mismatch in names list")
    total = 500
    names = ["Alice", 123, "Charlie"]  # 123 is not a string
    result = split_expense(total, names)
    print(result)
except TypeError as e:
    print(f"✅ TypeError caught: {e}\n")



```
